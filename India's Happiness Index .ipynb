```python
India = data[data["Country name"]=='India']
```

```python
India
```

```python
India.columns
```

```python
col_name=India[['Country name', 'year', 'Life Ladder', 'Log GDP per capita',
       'Social support', 'Healthy life expectancy at birth',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption', 'Positive affect', 'Negative affect']]
```

```python
col=col_name.corr()['Life Ladder']
```

```python
plt.figure(figsize=(15,10))
plt.suptitle("Features that contribute to the Life Ladder",size=20)
j=0
for i in col.index[1:]:
    ax=plt.subplot(5,2,j+1)
    ax=sns.lineplot(data=India,x='Life Ladder',y=i)
    ax.legend('')
    j=j+1
plt.legend('')
```

### Observation

The plots above show that high Social support 
and Perceptions of corruption have highly contributed in Happiness Index (Life Ladder) in India 

It can also be observed from the above graphs that Life Ladder increases as Log GDP per capita decreases

# Comparing India and Finland 

```python
newcolumns = data[['Life Ladder','Log GDP per capita',
       'Social support', 'Healthy life expectancy at birth',
       'Freedom to make life choices','Perceptions of corruption']]
col=newcolumns.corr()['Life Ladder']
```

```python
plt.figure(figsize=(10,12))
plt.suptitle("Camparing major effective features of India and Finland",size=20)
j=0
for i in col.index[1:]:
    
    
    ax=plt.subplot(5,1,j+1)
    ax=sns.lineplot(data=India,x='Life Ladder',y=i, linestyle='--')
    ax=sns.lineplot(data=Finland,x='Life Ladder',y=i)
    ax.legend('')
    j=j+1
plt.legend('')
```

### Conclusion

1. From The Graphs above we can conclude that by increasing Log GDP per capita, Social Support, High life expectancy at birth and Freedom to make life choices can increase the Happiness index of India
2. Increase the perceptions of corruptions will definitely increase the happiness index of India
3. Looking at the graphs we can say that if India remains at this path then by 5-6 years India will be higher in the Happiness index ladder
